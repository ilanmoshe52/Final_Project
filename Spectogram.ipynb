```
import matplotlib.pyplot as plt
import jax.numpy as jnp

def stft(a, n_fft=16, window=jnp.hanning):
  n = n_fft
  rpad = n - a.shape[-1] % n
  wins = jnp.pad(a, (0, rpad)).reshape(-1, n) * window(n)
  fftc = jnp.fft.fftshift(jnp.fft.fft(wins, n=n))[..., n // 2 : n]
  fftr = jnp.real(fftc * jnp.conj(fftc))
  return fftr

audio =  x
_ = plt.imshow(stft(audio, 64).T[:, -512:], cmap="viridis")
```
