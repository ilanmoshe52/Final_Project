```

import numpy as np
import matplotlib.pyplot as plt

# יצירת מערך נתונים עם רעש
np.random.seed(0)
data_with_noise = np.random.normal(0, 1, 1000)  # 1000 נקודות נתונים עם התפלגות נורמלית

plt.figure(1)
plt.plot(data_with_noise)
print(len(data_with_noise))

# הוספת ערכים קיצוניים
extreme_values = np.array([5, 6, 7, 8, 9, 10])
data_with_extremes = np.concatenate((data_with_noise, extreme_values))

plt.figure(2)
plt.plot(data_with_extremes)
print(len(data_with_extremes))

# חישוב האחוזונים
median = np.percentile(data_with_extremes, 50)  # אחוזון 50
quantile_90 = np.percentile(data_with_extremes, 90)  # אחוזון 90
quantile_95 = np.percentile(data_with_extremes, 95)

print(f"החציון (אחוזון 50) של הנתונים הוא: {median:.2f}")
print(f"האחוזון ה-90 של הנתונים הוא: {quantile_90:.2f}")
print(f"האחוזון ה-95 של הנתונים הוא: {quantile_95:.2f}")

# הדגמת השפעת הרעש על האחוזונים
noise = np.random.normal(0, 0.1, 1006)  # רעש קטן באותו גודל כמו המערך המקורי
data_with_more_noise = data_with_extremes + noise

# חישוב האחוזונים שוב אחרי הוספת הרעש
new_median = np.percentile(data_with_more_noise, 50)
new_quantile_90 = np.percentile(data_with_more_noise, 90)
new_quantile_95 = np.percentile(data_with_more_noise, 95)
print(f"החציון (אחוזון 50) של הנתונים עם רעש נוסף הוא: {new_median:.2f}")
print(f"האחוזון ה-90 של הנתונים עם רעש נוסף הוא: {new_quantile_90:.2f}")
print(f"האחוזון ה-95 של הנתונים הוא: {new_quantile_95:.2f}")
```
