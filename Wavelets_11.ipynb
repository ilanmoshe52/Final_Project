```

import pywt
import matplotlib.pyplot as plt

# Define the wavelet
wavelet = 'db10'

# Get the wavelet's filter coefficients
lpf, hpf = pywt.Wavelet(wavelet).filter_bank[:2]

# Time samples for plotting
samples = range(len(lpf))

# Plot the wavelet's low-pass filter
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.stem(samples, lpf, 'b', markerfmt='bo', basefmt='r-')
plt.title(f'Low-pass filter of the {wavelet} wavelet')
plt.xlabel('Coefficient number')
plt.ylabel('Amplitude')

# Plot the wavelet's high-pass filter
plt.subplot(3, 1, 2)
plt.stem(samples, hpf, 'g', markerfmt='go', basefmt='r-')
plt.title(f'High-pass filter of the {wavelet} wavelet')
plt.xlabel('Coefficient number')
plt.ylabel('Amplitude')

# Plot the wavelet itself
plt.subplot(3, 1, 3)
wavelet_function = pywt.Wavelet(wavelet)
phi, psi, x = wavelet_function.wavefun(level=5)
plt.plot(x, psi)
plt.title(f'{wavelet} wavelet')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
